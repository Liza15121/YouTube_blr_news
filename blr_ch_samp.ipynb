```python
def get_channel_stats (youtube, channel_ids):
    all_data = []
    request = youtube.channels().list(
        part='snippet,contentDetails,topicDetails,statistics',
        id=','.join(channel_ids)
    )
    response = request.execute()
    
    for i in range(len(response['items'])):
        data = {'channel_name': response['items'][i]['snippet']['title'],
                'channel_id': response['items'][i]['id'],
                'subscribers': response['items'][i]['statistics']['subscriberCount'],
                'views': response['items'][i]['statistics']['viewCount'],
                'total_videos': response['items'][i]['statistics']['videoCount'],
                'description': response['items'][i]['snippet']['description'],
                'playlist_id': response['items'][i]['contentDetails']['relatedPlaylists']['uploads'],
                'country': response['items'][i]['snippet'].get('country', 'NA'),
                'topicDetails': response['items'][i].get('topicDetails', {}),
                'publishedAt': response['items'][i]['snippet']['publishedAt'],
                
                }
        all_data.append(data)

        
    return all_data
```

```python
channels_data =get_channel_stats(youtube, channel_ids)
```

```python
df_channels_data = pd.DataFrame(channels_data)
```
